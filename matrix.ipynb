```python
a= [[1,2],
    [3,4]]

b = [[5,6],
     [7,8]]    

result = []

for i in range(len(a)): #row
    row=[]
    for j in range(len(a[0])):
        row.append(a[i][j] + b[i][j])
    result.append(row)    

print(result)
   
```

Let's see how we can multiply a matrix $A$ with a vector $\mathbf{v}$. We will multiply each row of matrix $A$ with the vector $\mathbf{v}$ to get a new vector.

Here, $A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$    and $\mathbf v = \begin{bmatrix}5 \\6 \end{bmatrix}$ 

```python
A = [[1, 2],
    [3, 4]]

v = [5, 6]

vector_multiply = [sum(A[i][j] * v[j] for j in range(len(v))) for i in range(len(A))]
vector_multiply
```

Next, let’s see how we can multiply two matrices $A$ and $B$, where

$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$ and $B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}$

```python
#i -> which row of A
#j-> which column of B
#k-> multiply element one by one

A = [[1, 2],
    [3, 4]]
B = [[5, 6],
     [7, 8]]

result = []
for i in range(len(A)): # row of A
    row = []
    for j in range(len(B[0])): #column of B
        sum_value = 0 
        for k in range(len(B)): #multiplication loop
            sum_value += A[i][k]*B[k][j]
        row.append(sum_value)
    result.append(row)  

print(result)          
```
