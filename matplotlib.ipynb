```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]

#plt.plot(x, y)
plt.title("Welcome to plot")
#print(type(plt.xlabel))
#plt.xlabel("cost")
#plt.ylabel("Item")

plt.plot(x,y,color= "red",marker="*",linewidth = 1,markersize = 6)
plt.show()
```

```python


plt.figure(figsize=(10,9))

x1 =[1,2,3,4,5]
x2 = [5,6,7,8,9]
x3 = [2,4,6,8,10]
x4 = [1,3,5,7,9]


plt.subplot(2,2,1)
plt.plot(x1,x2)
plt.title("First_Plot")

plt.subplot(2,2,2)
plt.plot(x2,x3)
plt.title("Second_Plot")

plt.subplot(2,2,3)
plt.plot(x3,x4)
plt.title("Third_Plot")

plt.subplot(2,2,4)
plt.plot(x3,x1)
plt.title("Fourth_plot")

plt.tight_layout()
plt.show()
```

###Bar Plot

```python

category = ['a','b','c','d','f','g']
value = [10,20,30,40,50,60]
plt.bar(category,value,color="purple")
```

##Histrogram 

```python
import matplotlib.pyplot as plt
data= [1,1,1,2,2,3,4,4,5,5,5,6,6,6,7,7,7,8,8,9,9,10]
plt.hist(data,bins=5)

```

##Scatter Plot

##Pie Chart

```python
label = ['a','b','c','d','e']
size = [30,50,60,30,40]

color = ["Red","blue","Green","yellow","orange"]

explode = (0.3,0,0,0,0)

plt.pie(size,labels=label,colors=color,explode=explode)
```
