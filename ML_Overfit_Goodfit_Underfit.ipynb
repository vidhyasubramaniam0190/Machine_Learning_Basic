Underfitting : MOdel too simple
Goodfitting : Balance model
Overfitting : Model too complex(memorize noise)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

np.random.seed(0)
X = np.sort(np.random.rand(30,1),axis = 0) 
       # creating 30 row 1 column  axis = 0 -> sort column-wise  -> top to bottom
       # axis = 1 -> sort row-wise  - left to right
       # x values are random so if we not sort graph look messy [ need multiple columns]
       # sort -> arranging the value
       # axis -> direction
y = np.sin(2*np.pi*X).ravel()

y=y+np.random.normal(0,0.2,y.shape)

def plot_model(degree,title):
    model = make_pipeline(PolynomialFeatures(degree),
    LinearRegression())
    model.fit(X,y)

    X_test = np.linspace(0,1,100).reshape(-1,1)
    y_pred = model.predict(X_test)

    plt.scatter(X,y,color='black')
    plt.plot(X_test,y_pred)
    plt.title(title)
    plt.show()

   # Underfitting (degree = 1)
plot_model(1, "Underfitting (Simple Model)")

# Good Fit (degree = 4)
plot_model(4, "Good Fit")

# Overfitting (degree = 15)
plot_model(15, "Overfitting (Complex Model)")

```
