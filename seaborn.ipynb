```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
xArray = [50,60,70,80,90,100,110,120,130,140,150];
yArray = [7,8,8,9,9,9,10,11,14,14,15];

data = pd.DataFrame({
    "xArray": xArray,
    "yArray": yArray
})
data

sns.scatterplot(x="xArray",y="yArray",data=data)
plt.title("Heading")
plt.show()

```

## LINE PLOT

```python
sns.lineplot(x="xArray",y="yArray",data=data)
plt.title("Line plot")
plt.show()
```

```python
sns.barplot(x="xArray",y="yArray",data=data)
plt.title("bar plot")
plt.show()
```

```python
sns.histplot(x="xArray",y="yArray",data=data)
plt.title("Histrogram plot")
plt.show()
```

```python
sns.boxplot(x="xArray",y="yArray",data=data)
plt.title("Histrogram plot")
plt.show()
```

```python
csv_file = pd.read_csv(r"C:\Users\Vidhya\Python_Program\people-100.csv")
csv_file

import random
csv_file["Salary"] = [random.randint(10000,99999)
for _ in range(len(csv_file))]
csv_file["Apprasial"] = csv_file["Salary"].apply(lambda X:X*2)
csv_file




```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = csv_file.head(10)

sns.barplot(x="Region", y="Apprasial", data=data)

plt.title("Region vs Appraisal")
plt.show()
```

```python
sns.boxplot(x="Region", y="Apprasial", data=data)
plt.title("Boxplot")
plt.show()
```
