This is not vector addition so let we try for vector


```python
def add_vector(v1,v2):
    if len(v1)!= len(v2):
        raise ValueError("vector must be same character")
    return tuple(v1[i] + v2[i] for i in range(len(v1)))

add_vector(v1,v2)   
```

#subtract the Vector

```python
def sub_vector(v1,v2):
    if len(v1)!= len(v2):
        raise ValueError("vector must be same character")
    return tuple(v1[i] - v2[i] for i in range(len(v1)))

sub_vector(v1,v2)   
```

When adding two vectors like (3, 2, 5) and (4, 5, 6), using the '+' operator doesn’t work the way you would expect, it just joins the tuples instead of adding the corresponding components. That’s why we define a function that adds elements one-by-one.

```python
def vector_add(v1, v2):
    return tuple(a + b for a, b in zip(v1, v2))
```

Subtraction works the same way as addition, where we define a function that subtracts elements one-by-one.

```python
def vector_subtract(v1, v2):
    return tuple(a - b for a, b in zip(v1, v2))
```

For the dot product, we multiplied the matching elements and added the results.

```python
dot_product = 0

for i in range(len(v1)):
    dot_product += v1[i] * v2[i]

print("Dot Product:", dot_product)
```

However, there are situations where you might want to modify the vector, such as scaling a vector by multiplying each of its components by a number, which is useful when you want to make a vector longer or shorter without changing its direction. In these cases, lists may be more appropriate because they allow in-place modifications.

```python
v = [1, 2, 3]
k = 2
for i in range(len(v)):
    v[i] *= k
```

You also can do this using tuple, but since it is immutable, the scaling function will return a new tuple instead of modifying the original one.

```python
v = (1, 2, 3)
k = 2

def scale(v, k):
    return tuple(k * a for a in v)
scale(v,k)
```

```python
def linear_combination2(vectors, coefficients):
  # Check if all vectors have the same dimension
  num_dimensions = len(vectors[0])
  for vector in vectors:
    if len(vector) != num_dimensions:
      print('Dimensions are not consistent')  
      break

  # If dimensions are consistent, proceed with the calculation
  result = [0] * num_dimensions
  for i in range(num_dimensions):
    for j, vector in enumerate(vectors):
      result[i] += coefficients[j] * vector[i]
  return tuple(result)
```

```python
v1 = (3, 2, 5)
v2 = (4, 6, 1)
coefficients = [2, 3]

result = linear_combination2([v1, v2], coefficients)
result 
```

Hari, Sujata and Jeevan are fond of reading books. They rate a set of five books that they have read. Hari’s ratings are represented using h = (3,4,3,3,5), while Sujata’s ratings and Jeevan’s ratings are represented using s= (2,4,5,4,4) and j=(4,2,2,3,1), respectively. Whose preferences are similar to Sujata’s?

```python
import numpy as ny
h = ny.array([3,4,3,3,5])
s= ny.array([2,4,5,4,4])
j= ny.array([4,2,2,3,1])

# Distance from sujatha
dis_h = ny.linalg.norm(h-s)
dis_j = ny.linalg.norm(j-s)

print(dis_h)
print(dis_j)
```

Mahi and Dhruv are preparing for an entrance exam. Their marks in three subjects in the first term exam are represented using the vectors m1 =(12,19,25) and d=(20,22,25), and their marks in the same three subjects in the second term exam are represented using the vectors m2=(20,20,27) and d2= (20,20,25) . The improvement in Mahi’s score can be calculated using L2 norm (m1-m2). Similarly, Dhruv’s improvement is given by L2 norm (d1-d2). Who has shown more improvement: Dhruv or Mahi?

```python
m1 =ny.array([12,19,25])
d1 = ny.array([20,22,25])

m2=ny.array([20,20,27])
d2 = ny.array([20,20,25])

m_norm = ny.linalg.norm(m1-m2)
d_norm = ny.linalg.norm(d1-d2)

print(m_norm)
print(d_norm)
```
